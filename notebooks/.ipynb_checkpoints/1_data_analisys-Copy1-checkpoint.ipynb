## Geographic Location x Price

Lets consider the houses's latitude and longitude impact on price.

```python
figure(figsize=(12, 7.2))
plt.scatter(df['longitude'], df['latitude'], cmap='Reds', c=df['price'], s=10)
plt.colorbar().set_label('Price')
plt.clim(vmin=0, vmax=1500000)
plt.title('Location x Price')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show();
```

Latitude seems to play a important role, as houses in higher latitudes tend to be more expensive (slightly dark tone). 

The longitude itself does not seem to matter much, except for its relation with the house's proximity to the margin of some confined blank areas. For similar longitudes, houses by these margins are generally more expensive, while those between them dont vary much in price. 

Assuming that some of these blank areas are water (they could also be public spaces or green areas), this would help explain the longitude's lower correlation and the waterfront status's higher correlation with the price tag. 

Lets now check the waterfront status.

```python
figure(num=None, figsize=(8, 5), dpi=80, facecolor='w', edgecolor='k')
ax = sns.boxplot(x='is_waterfront', y='price', data=df, showfliers=False)
ax.set(title='Waterfront x House price', xlabel='Waterfront', ylabel='Price');
ax = plt.xticks(rotation=90);
```

Indeed waterfront houses are generally much more expensive, which validates our first graph and confirm this feature's correlation with the price tag.

### Average size of neighbor houses

Now lets check the impact of the average size of nearby houses on the price tag.

```python
figure(num=None, figsize=(8, 5), dpi=80, facecolor='w', edgecolor='k')
ax = plt.scatter(df['avg_size_neighbor_houses'], df['price'], s=10)
plt.title('Average size of neighbor houses x Price')
plt.xlim(0, 6000)
plt.xlabel('Average size')
plt.ylabel('Price')
plt.show();
```

```python
print('Pearson Coefficient: %.4f' % df.corr().loc['avg_size_neighbor_houses', 'price'])
figure(num=None, figsize=(8, 5), dpi=80, facecolor='w', edgecolor='k')
ax = sns.regplot(x='avg_size_neighbor_houses', y='price', data=df, scatter_kws={'s':2})
ax.set(title='Avg. Neig. Houses Size x Price', xlabel='Average neighbor houses size', ylabel='Price');
```

As we can see, the impact, on price, of the average size of neighbor houses and of the size of the house itself is very similar. 

Possibly, the average size of neighbor houses is too good of an estimator for the size of the house, i.e. they are highly correlated. In this scenario, the average value would, ultimately, for the purposes of our model, work as a second representation of the house size. 

Lets explore their relation further by building a regression graph.

```python
print('Pearson Coefficient: %.4f' % df.corr().loc['avg_size_neighbor_houses', 'size_house'])
figure(num=None, figsize=(8, 5), dpi=80, facecolor='w', edgecolor='k')
ax = sns.regplot(x='avg_size_neighbor_houses', y='size_house', data=df, scatter_kws={'s':2})
ax.set(title='Avg. Neig. Houses Size x House Size', xlabel='Average neighbor houses size', ylabel='House size');
```

Indeed they are highly correlated, with their Pearson coefficient being 0.75 (1 meaning total positive linear correlation).

This is not great for the prediction model as it effectively means less relevant information due to data redundancy. It increases training and prediction times and may lead to overfitting. 

So it is worth considering using only the one most correlated with the target variable, the house size in this case.

### Number of Bathrooms
The number of bathrooms is represented in intervals of 0.25, with the minimum being 0 and the maximum being 8. It would be wise to check if this was intended or if non-integers values are missrepresentations. 

For instance, a house with 1.5 bathrooms could mean that it has a normal size bathroom and a very small one, or  that it has a full fledged bathroom and another one with just a toiled seat and a lavatory.

Lets study this feature's correlation with house prices to make a decision.

```python
figure(num=None, figsize=(8, 5), dpi=80, facecolor='w', edgecolor='k')
ax = sns.boxplot(x='num_bath', y='price', data=df, showfliers=False)
ax.set(title='Bathrooms x House price', xlabel='Number of bathrooms', ylabel='Price');
ax = plt.xticks(rotation=90);
```

From the graph above, we can see that the number of bathrooms has a positive correlation with the house price. Since houses with more toilets are more valued by buyers, this makes sense. 

The number of bathrooms is represented in intervals of 0.25, with the minimum being 0 and the maximum being 8.
Non-integer values could be a missrepresentation but, as they maintain the patter of their adjacent ones, we can assume that the data was well measured and that this way of representation was intended.

For instance, a house with 1.5 bathrooms could mean that it has a normal size bathroom and a very small one, or  that it has a full fledged bathroom and another one with just a toiled seat and a lavatory.
