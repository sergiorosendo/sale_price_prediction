## House qualities exploration

### House size

```python
plot_size(df)
```

As expected, bigger houses tend to have higher prices. 

### Average size of neighbor houses

```python
plot_avg_size_neigh(df)
```

Larger neighbor houses tends to increase the house price. This can mean that richer neighborhoods maker for pricier homes.

Also, this feature's impact on price can come from its relation to the house size itself. Homes in the same neighborhood often have similar size, as indicated by the high Pearson correlation between these variables.

```python
from scipy.stats import pearsonr

print('Pearson corr. with house size: %.2f' % pearsonr(df['size_house'], df['avg_size_neighbor_houses'])[0])
```

### Number of Bathrooms

```python
plot_bathrooms(df)
```

The number of bathrooms has a positive correlation with the house price. Since houses with more toilets are more valued by buyers, this makes sense. 

This feature is represented in intervals of 0.25, with the minimum being 0 and the maximum being 8.
Non-integer values could be a missrepresentation but, as they maintain the pattern of their adjacent ones, we can assume that the data was well measured and that this way of representation was intended.

A house with 1.5 bathrooms could, for instance, mean that it has a normal size bathroom and a very small one, or mean that it has a full fledged bathroom and another one with just a toiled seat and a lavatory.

### Number of Floors

```python
plot_floors(df)
```

Homes with only 1 floor tend to be cheaper, increasing in price with floors up to 2.5. With noticeable lesser price quantiles for homes above 2.5 floors.

### Condition

```python
plot_condition(df)
```

Houses in conditions 1 and 2 are much cheaper than the others. Houses with condition 3, 4 and 5 have very similar price quantiles.

This may indicate its important to keep the quality of the house aboce a certain threshold (in this case 2). Past that point, spending more resources to repair and maintain it, for the purpose of selling, may be a waste of money.
