### Average size of neighbor houses

Now lets check the impact of the average size of nearby houses on the price tag.

```python
from src.visualization.visualize import plot_avg_size_neigh

plot_avg_size_neigh(df)
```

As we can see, the impact, on price, of the average size of neighbor houses and of the size of the house itself is very similar. 

Possibly, the average size of neighbor houses is too good of an estimator for the size of the house, i.e. they are highly correlated. In this scenario, the average value would, ultimately, for the purposes of our model, work as a second representation of the house size. 

Lets explore their relation further by building a regression graph.

```python
print('Pearson Coefficient: %.4f' % df.corr().loc['avg_size_neighbor_houses', 'size_house'])
figure(num=None, figsize=(8, 5), dpi=80, facecolor='w', edgecolor='k')
ax = sns.regplot(x='avg_size_neighbor_houses', y='size_house', data=df, scatter_kws={'s':2})
ax.set(title='Avg. Neig. Houses Size x House Size', xlabel='Average neighbor houses size', ylabel='House size');
```

Indeed they are highly correlated, with their Pearson coefficient being 0.75 (1 meaning total positive linear correlation).

This is not great for the prediction model as it effectively means less relevant information due to data redundancy. It increases training and prediction times and may lead to overfitting. 

So it is worth considering using only the one most correlated with the target variable, the house size in this case.
