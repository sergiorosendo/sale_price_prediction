### Number of Bathrooms

```python
from src.visualization.visualize import plot_bathrooms

plot_bathrooms(df)
```

The number of bathrooms has a positive correlation with the house price. Since houses with more toilets are more valued by buyers, this makes sense. 

This feature is represented in intervals of 0.25, with the minimum being 0 and the maximum being 8.
Non-integer values could be a missrepresentation but, as they maintain the pattern of their adjacent ones, we can assume that the data was well measured and that this way of representation was intended.

A house with 1.5 bathrooms could, for instance, mean that it has a normal size bathroom and a very small one, or mean that it has a full fledged bathroom and another one with just a toiled seat and a lavatory.
